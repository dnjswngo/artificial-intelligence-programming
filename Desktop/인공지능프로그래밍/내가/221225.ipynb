기계학습: 데이터들로부터 학습하여 스스로 발전할 수 있는 컴퓨터 기술 및 분야
>컴퓨터가 이해할 수 있는 데이터 형태를 특징점

표현학습: 데이터들로부터 특징점 추출을 자동으로 학습하는 기계 학습의 일정
>한 단계만에 특징점으로 변환

딥러닝: 표현 학습 및 인공 신경망에 기반하여 계층적으로 특징점 추출을 학습하는 기계 학습의 일종
> 사물인식, 음성인식, 텍스트 이해 등 수행

인공신경망: 실제 뇌의 신경세포, 신경망 구조를 기반으로 이와 유사한 구조를 갖도록 만든 기계 학습 방식의 일종


low-level feature : 사람에게 익숙하지 않고 컴퓨터 입장에서 익숙하고 수치로 나타내기 좋은 특징점
이미지를 픽섹 하나하나가 눈에 보일만큼 확대해서 분석한 특징점
high-level feature : 사람에게 익숙하지만 컴퓨터 입장에서 이해, 수치화 어려움
이미지 전체 모양, 질감 등에 해당함

저레벨 특징점에서 고레벨 특징점 추출까지 학습하는 것 ~~ 딥러닝


지도학습: 컴퓨터에게 각 데이터에서의 정답(label)이 무엇인지 알려주고 학습시키는 방법
>분류(주어진 데이터가 어떤 class에 속하는지를 예측), 회귀(주어진 입력 데이터로부터 원하는 결과값 자체를 예측)

비지도학습:데이터만 주어지고 정답x, 주로 어떤 데이터들끼리 하나의 집합으로 묶일 수 있는지를 학습 

강화학습: 데이터가 주어지는 것이 아니라 머신러닝 모델이 직접 행동을 하며 각각의 행동에 대해 잘했는지 못했는지 피드백을 받으며 학습하는 방식

## 선형회귀
회귀란 우리가 얻을 수 있는 독립변수들로부터 우리가 알고자 하는 결과값을 도출해내는 모델을 얻는 머신러닝 기법

회귀에서의 레이블은 연속된 범위를 갖고 있다고 볼 수 있음

이 때 데이터와 레이블의 관계를 선형으로 가정하고 문제를 푸는 방식을 선형 회귀

머신러닝에서 데이터와 레이블 사이의 관계에 대해 가정한 사항들을 hypothesis이라고함.

선형 회귀에서는 데이터와 레이블 사이에 선혀 관계가 있다고 가정하며 따라서 일반적으로 아래와 같은 형해의 가설을 수립함
>y=Wx+b

>w는 weight b는 bias >>선형회귀의 목표는 주어진 학습 데이터에 가장 알맞은 w, b 찾아내는 것

주어진 학습 데이터에 가장 알맞는다는 것은 어떤 기준을 세우느냐에 따라 달라질 수 있음.(비용함수 손실함수 오차함수 목적함수)



MSE(평균 제곱 오차): 널리 쓰이는 손실 함수 중 하나, MSE값을 최소로 하는 W, b의 값을 찾는 것이 선형 회귀의 목표

경사하강법(gradient descent): W, b 값을 조금씩 변경해가며 손실 함수 값이 더 작은 값을 찾아가는 방식, 변경하는 방향은 미분을 통해 알 수 있으며, 학습률이라는 별도의 값을 통해서 한번에 변경하는 값의 양을 결정함

선형 회귀의 경우 경사하강법 대신 행렬 연산으로 바로 최적값을 구할 수 있으나, 더욱 복잡한 머신러닝 모델들에 대해서는 경사하강법이 필수

```python
import torch as t
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
t.manual_seed(1)
```

```python
#데이터셋 생성
x_train= t.FloatTensor([[1],[2],[3]])
y_train= t.FloatTensor([[2],[4],[6]])
#가중치 편향 생성
W=t.zeros(1, requires_grad=True)
b=t.zeros(1, requires_grad=True) #autograd 가능한 변수
#가설 수립
hypo=x_train*W+b
#비용함수 선언
cost =t.mean((hypo-y_train)**2)
```

최적화 담당 객체를 optimizer라고 함

SGD(stochastic gradient descent): 기본적인 optimizer, 최적화 수행할 대상 변수들을 첫 번째 매개변수로 입력하고, 학습률을 lr에 저장.

이후

zero_grad(): 미분값 초기화

backward(): 현재 손실 함수값을 바탕으로 미분값 계산

step(): 현재 미분값 및 학습률을 바탕으로 변수 업데이트 1회 수행

얘네로 최적화 수행

```python
x_train=t.FloatTensor([1,2,3]).view(-1,1)
y_train=t.FloatTensor([2,4,6]).view(-1,1)

optimizer=optim.SGD([W, b], lr=0.01) #optimizer 설정, 학습률 0.01
nb_epochs=1000 #원하는만큼 경사 하강법을 반복
#epoch 학습 데이터셋 전체를 한번 씩 다 학습하는 반복 단위

for epoch in range(nb_epochs+1):
  #가설 및 손실 함수 위에서 정의함
  hypo=x_train*W+b
  cost =t.mean((hypo-y_train)**2)

  #경사하강법
  optimizer.zero_grad() #미분값 초기화
  cost.backward() #현재 손실 함수 값 바탕으로 미분하는거
  optimizer.step() #이걸로 W, b 값 바뀜> 변수 업데이트

  #100번마다 로그 출력
  if epoch %100==0:
    print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(epoch, nb_epochs, W.item(), b.item(), cost.item()))
  
```

학습하는동안
>현재 가중치, 편향을 이용한 예측값 도출,
예측값과 레이블을 이용한 손실값 도출
미분값 초기화, 미분값 계산, W b값 업데이트

미분값을 초기화하지 않고 계산하면 계속해서 미분값들이 쌓이게 되어 있으므로 업데이트에서 사용한 후에는 미분값을 초기화해주어야 함

```python
test_data=t.FloatTensor([[4]])
predic=test_data*W + b
print(predic)
```

## 다중선형회귀


```python
x_train=t.FloatTensor([[73,80,75],[93,88,93],[89,91,80],[96,98,100]])
y_train=t.FloatTensor([[152],[185],[180],[196]])

W=t.zeros((3,1), requires_grad=True)
b=t.zeros(1, requires_grad=True)

optimizer=optim.SGD([W,b], lr=1e-5)

nb_epochs=1000
```

```python
for epoch in range(nb_epochs+1):
  hypo=x_train.matmul(W)+b
  cost=t.mean((hypo-y_train)**2)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  w_list=W.view(-1).tolist()
  if epoch%100==0:
    print('Epoch {:4d}/{} w1: {:.3f}, w2: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(epoch, nb_epochs, w_list[0], w_list[1], b.item(), cost.item()))
```
